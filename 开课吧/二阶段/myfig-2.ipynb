```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.arange(0,1,0.05)
print(x)
```

```python
# y=sin(2*pi*x)
y=np.sin(2*np.pi*x)
print(y)
```

```python
plt.plot(x,y,'b--*',label='sinx')
plt.title('sinx')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.show()
```

```python
fig,ax=plt.subplots(2,2)
ax[0,1].plot(x,y)
plt.show()
```

```python
y2=np.cos(2*np.pi*x)
print(y2)
```

```python
fig,ax=plt.subplots()
ax.plot(x,y,label='sin')
ax.plot(x,y2,label='cos')
ax.set(title='cos sin')
ax.legend(loc='best')
plt.show()
```

```python
# 
fig.savefig('myfig.png')
```

```python
#
import numpy as np
data=np.random.randint(0,100,size=[40,40])
print(data)
```

```python
# 
x,y=data[0],data[1]  #

ax=plt.subplot(111)
ax.scatter(x[:10],y[:10],c='r')
ax.scatter(x[10:20],y[10:20],c='b')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

```python
#
from mpl_toolkits.mplot3d import Axes3D
x,y,z=data[0],data[1],data[2]
ax=plt.subplot(111,projection='3d')
ax.scatter(x[:10],y[:10],z[:10],c='r')
ax.scatter(x[10:20],y[10:20],z[10:20],c='b')
ax.scatter(x[30:40],y[30:40],z[30:40],c='g')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
