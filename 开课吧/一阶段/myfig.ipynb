```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.arange(0,1,0.05)
print(x)
```

```python
# y=sin(2*pi*x)
y=np.sin(2*np.pi*x)
print(y)
```

```python
plt.plot(x,y,'b--*',label='sinx')
plt.title('sinx')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.show()
```

```python
fig,ax=plt.subplots(2,2)
ax[0,1].plot(x,y)
plt.show()
```

```python
y2=np.cos(2*np.pi*x)
print(y2)
```

```python
fig,ax=plt.subplots()
ax.plot(x,y,label='sin')
ax.plot(x,y2,label='cos')
ax.set(title='cos sin')
ax.legend(loc='best')
plt.show()
```

```python
# 
fig.savefig('myfig.png')
```
