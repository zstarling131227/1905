```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.arange(0,1,0.05)
print(x)
```

```python
# y=sin(2*pi*x)
y=np.sin(2*np.pi*x)
print(y)
```

```python
plt.plot(x,y,'b--*',label='sinx')
plt.title('sinx')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.show()
```

```python
fig,ax=plt.subplots(2,2)
ax[0,1].plot(x,y)
plt.show()
```

```python
y2=np.cos(2*np.pi*x)
print(y2)
```

```python
fig,ax=plt.subplots()
ax.plot(x,y,label='sin')
ax.plot(x,y2,label='cos')
ax.set(title='cos sin')
ax.legend(loc='best')
plt.show()
```

```python
# 保存图片
fig.savefig('myfig.png')
```

```python
# 读取数据
import pandas as pd
# df=pd.read_csv('/home/tarena/tarena/1905/开课吧/一阶段/data.csv')
df=pd.read_csv('../一阶段/data.csv',index_col='年份') # 将年份设置为索引，默认为0，---n
df.head()
```

```python
x=df.index.values  # 获取年份索引
y=df['人均GDP（元）'].values # 获取人均GDP
print(x,y)
```

```python
from pylab import mpl

mpl.rcParams['font.sans-serif']=['FangSong']  # 指定默认字体（ｗｉｎｄｏｗ系统）
```

```python
from matplotlib import font_manager
# my_font = font_manager.FontProperties(fname = '/usr/share/fonts/arial/ARIALNB.TTF') # 指定默认字体(不能用中文)
my_font = font_manager.FontProperties(fname = '/usr/share/fonts/song/SIMLI.TTF') # 指定默认字体（宋体）
# my_font = font_manager.FontProperties(fname = '/usr/share/fonts/fangsong/SIMFANG.TTF') # 指定默认字体（仿宋）

fig,ax=plt.subplots()
ax.plot(x,y,'r--*')
# ax.set(title='人均GDP走势图',xlabel='年份',ylabel='人均GDP')  # 设置标题，但是需要设置字体格式，否则会乱码
#　修改字体，使标签可以用中文显示
plt.title('人均GDP走势图',fontproperties = my_font)
plt.xlabel('年份',fontproperties = my_font)
plt.ylabel('人均GDP',fontproperties = my_font)
plt.show()
```

```python
fig,ax=plt.subplots()
# ax.pie(y)
ax.pie(y[:5],labels=x[:5])
plt.show()
```

```python
fig,ax=plt.subplots()
# ax.pie(y)
ax.pie(y[:5],labels=x[:5],explode=[0,0.05,0.1,0.15,0.2]) # explod 设置的是饼块之间的间距
plt.show()
```
