```python
data=pd.read_csv('hotel_data.csv',encoding='gb2312',index_col='Unnamed: 0')
data.head()
```

```python
#每个地区的评分人数的总和以及均值
data["评分人数"]=data["评分人数"].astype("int")
data.评分人数.groupby(data["地区"]).agg([np.sum,np.mean])
```

```python
#获取index和columns
data.columns
data.index
```

```python
data.combine_first
data.columns
```

```python
data.head()
```

```python
#每个酒店类型的平均价格、最高价、最低价

data["价格"]=data["价格"].astype("int")
data.价格.groupby(data["类型"]).agg([np.mean,np.max,np.min])
```

```python
data.head()
```

```python
# 数据离散化，按照价格将酒店分为3个等级,0-500为C，500-1000为B,大于1000为A，
# 列名设置为“价格等级”。

data["价格等级"]=pd.cut(data["价格"],bins=[0,500,1000,50000],labels=["C","B","A"])
data.head()
```

```python
#获取评分均值最高和最低的地区的数据

data["评分"]=data["评分"].astype("int")
group=data.评分.groupby(data["地区"]).mean()

group=group.sort_values(ascending=False)
group
```

```python
group.index
```

```python
avg_max=data[data["地区"].isin(group.index[:8])]
avg_min=data[data["地区"]=="油尖旺"]
```
